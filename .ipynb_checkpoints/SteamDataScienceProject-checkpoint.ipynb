```python
# Import of pandas and numpy
import pandas as pd
import numpy as np

# Import of seaborn and matplotlib
import seaborn as sns 
import matplotlib.pyplot as plt

# Import from sklearn the linear_model
from sklearn import datasets, linear_model
```

```python
# In here we are getting the enconder
import chardet
file = 'U:\SteamDataScienceProject\Valve_Player_Data.csv'
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
# result
```

```python
# In here we are retreaving the csv's info and getting in into a DataFrame
playerData = pd.read_csv(file, encoding='ISO-8859-1')

# To check all of the NaN values inside the dataframe
# playerData[playerData.isna().any(axis=1)]

# It is needed to drop the NaN values not get the anaylisis influenced by them
playerData = playerData.dropna()
playerData
```

```python
# It is needed to drop the na values just in case to avoid errors
playerData.dropna(inplace = True)
```

```python
# In here we are creating a subset, getting rid of the URL column
data = playerData[["Game_Name", "Avg_players", "Gain",
                     "Percent_Gain", "Peak_Players", "Date"]]
# usable
```

```python
# We are setting the right format, this being yy-mm-dd
data["Date"] = pd.to_datetime(data["Date"], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')
# usable
```

```python
# Obtain rows of the dataframe based on their date
data2021 = data.loc[(data['Date'] >= '2021-01-01')]
```

```python
# Check if there are any nulls
# data2021.isnull().sum()
```

```python
# Order in which things are being done here: 
# 1. I am creating a subset that only includes game_name and Avg_players
# 2. I am grouping them by Game_Name and getting the mean of each
# 3. I am sorting them by Avg_Players, from biggest to smaller
data2021Players = data2021[["Game_Name", "Avg_players"]].groupby(['Game_Name']).mean().sort_values(by='Avg_players', ascending=False)
# data2021Players
```
