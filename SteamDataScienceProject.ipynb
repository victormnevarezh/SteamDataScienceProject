```python
# Import of pandas and numpy
import pandas as pd
import numpy as np
```

```python
# In here we are getting the enconder
import chardet
file = 'U:\SteamDataScienceProject\Valve_Player_Data.csv'
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
# result
```

```python
# In here we are retreaving the csv's info and getting in into a DataFrame
playerData = pd.read_csv(file, encoding='ISO-8859-1')
# playerData
```

```python
# In here we are creating a subset, getting rid of the URL column
usefulData = playerData[["Game_Name", "Month_Year", "Avg_players", "Gain", "Percent_Gain", "Peak_Players", "Date"]]
usefulData
```

```python
# Replace
usefulData.replace()
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-21', 'May 2021', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-20', 'May 2020', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-19', 'May 2019', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-18', 'May 2018', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-17', 'May 2017', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-16', 'May 2016', inplace=True)
usefulData['Month_Year'].mask(usefulData['Month_Year'] == 'may-15', 'May 2015', inplace=True)
usefulData
```
